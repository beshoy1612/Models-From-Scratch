```python
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))
```

```python
def log_loss(y_true, y_predicted):
    epsilon = 1e-15
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
class nural:
    def __init__(self):
        self.w = 1
        self.b = 0
        self.lr = 0.01

    def fit(self, X_train, y_train, epochs):
        self.w, self.b = self.GD(X_train, y_train, epochs, threshold=0.1)

    def predict(self, X_test):
        y_pred = sigmoid((np.dot(X_test, self.w)) + self.b)
        y_pred = [1 if i >=0.5 else 0 for i in y_pred]
        return y_pred

    def GD(self, X_train, y_train, epochs, threshold):
        n_sample, n_feature = X_train.shape
        w = np.zeros(n_feature)
        b, lr, threshold = 0, 0.01, 0.1
        for _ in range(epochs):
            linear = np.dot(X_train, w) + b
            #  linear ypred loss updtae
            y_pred = sigmoid(linear)
            loss = log_loss(y_train, y_pred)
            dw = 1 / n_sample * np.dot(np.transpose(X_train), (y_pred - y_train))
            db = 1 / n_sample * np.sum(y_pred - y_train)
            w -= lr * dw
            b -= lr * db
            if threshold >= loss:
                break

        return w, b
    
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/deep-learning-keras-tf-tutorial/refs/heads/master/6_gradient_descent/insurance_data.csv')
```

```python
from sklearn.model_selection import  train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```
