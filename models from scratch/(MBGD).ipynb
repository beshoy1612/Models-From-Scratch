```python
import numpy as np
```

```python
def MBGD(X_train, y_train, epochs):
    n_sample, n_feature = X_train.shape
    w = np.zeros(n_feature)
    b, lr = 0, 0.01
    for _ in range(epochs):
        indicies = np.random.choice(len(X_train), size=5, replace=False)
        X_sample, y_sample = X_train[indicies], y_train[indicies]
        y_pred = np.dot(X_sample, w) + b

        dw = 1 / len(X_sample) * np.dot(np.transpose(X_sample), (y_pred - y_sample))
        db = 1 / len(X_sample) * np.sum(y_pred - y_sample)
        w -= lr * dw
        b -= lr * db

    return w, b
```
