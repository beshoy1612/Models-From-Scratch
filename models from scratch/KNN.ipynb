```python
import numpy as np
from collections import Counter
```

```python
def Euclidean(x,y):
    return np.sqrt(np.sum((x-y)**2))
```

```python
class KNN:
    def __init__(self,k=3): # define default value for K
        self.k = k

    def fit(self,x,y): # create two varaible to use it in predicit 
        self.x_train=x
        self.y_train=y

    def predict(self,X_test): # take each point in X_test to calculate distance between it and other points
        result=[self._predict(i) for i in X_test] # call _predict() function to claculate distance
        return np.array(result) # return result to array
    
    def _predict(self,X):
        distance=[Euclidean(x,X) for x in self.x_train] # call Euclidean function to claculate distance
        sort_distance=np.argsort(distance)[:self.k] # sort distance to show the K nearest point  
        label=[self.y_train[i] for i in sort_distance] # give point label to cluster it
        common=Counter(label).most_common(1) # show the nearst point
        return common[0][0] # return the label name from the pair
```

#### test my model

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X, y = iris.data, iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
model=KNN()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
acc = np.sum(y_pred==y_test)/len(y_test)
```
